# Problem 2: Analyzing code running times (33 points [4,4,6,6,5,8])

Let N = len(L), where L is the input to each function.  Find the worst case big-$\Theta$ asymptotic complexity of the following functions.

````python
# (a)
# A function that sets each element of a list to the list average
def set_to_avg(L):
    my_sum = 0
    count = 0
    for item in L:
        my_sum += item
        count += 1
    for i in range(len(L)):
        L[i] = my_sum/count
    return L
````

a, 4 points) 2N --> $\Theta(N)$

````python
# (b)
# A function that finds all possible sums of 3 different
# numbers from the list, returning the sums as a set
def all_sums_of_two(L):
    sums = set()
    for i in L:
        for j in L:
            # Hint:  This "if" doesn't really rule out
            # enough items to make us worry about its effect
            # on the running time
            if i != j:
                sums.add(i+j)
    return sums
````

# Problem 3:  Analyzing algorithms (5 points)

Consider the following sorting algorithm that tries to have it all.  First, it tries to run insertion sort, but it keeps a tally of the number of shift operations it performs.  If the number of operations exceeds $n \log N$ (which you can assume doesn't happen if the array is already sorted), it gives up on insertion sort, and runs mergesort instead on whatever state the array was left in.

What is the best-case running time of this algorithm (in big-O), and what is the worst-case time?  Explain your reasoning.

The best case of the algorithm is O(N) which would occur if the array is already sorted. The worst case running time would be if it exceeded N log N and had to switch to mergesort. It is possible that this would happen because the worst case for insertion sort is O(N^2) which is greater than N log N. Therefore if it is the worst case, the algorithm would hit N log N before the sort was completed and then have to continue with mergesort. The worst case of merge sort is O(N log N), meaning the worst case of this algorithm is O(N log N).

# Problem 4:  Graphs and Centrality (20 pts)

a, 8 pts) The following is a representation of an undirected graph with an adjacency list representation.  Write a function friends_of_friends() that can take a vertex name and an arbitrary graph like this as arguments, and return a list of all vertices that are distance 2 from the first argument.  (Do not return vertices that are distance 1 or 0, and do not repeat any vertices.)  For example, on this graph, friends_of_friends('A', graph) should return ['D', 'E'], because 'B' and 'C' are distance 1, 'D' has a distance 2 path through both, and 'E' has a distance 2 path through 'C'.

````python
graph = {'A':  ['B','C'],
         'B':  ['A','C','D'],
         'C':  ['A','B','D','E'],
         'D':  ['B','C'],
         'E':  ['C','F'],
         'F':  ['E']}
````

````python
def friends_of_friends(vertex, graph):
  answer = []
  for node in graph[vertex]:
    for i in graph[node]:
      if i != vertex and i not in graph[vertex] and i not in answer:
        answer.append(i)
  return answer
````

````python
print(friends_of_friends('A', graph)) # expect ['D', 'E']
print(friends_of_friends('B', graph)) # expect ['E']
print(friends_of_friends('C', graph)) # expect ['F']
print(friends_of_friends('D', graph)) # expect ['A', 'E']
print(friends_of_friends('E', graph)) # expect ['A', 'B', 'D']
````

b, 12 pts) Write a function closeness_centrality() that calculates *approximate* closeness centrality, given a vertex name and a graph.  You can assume any distances that are greater than 2 are roughly 3 - with this assumption, you won't need a full breadth first search, but you can instead use your friends_of_friends() function to determine which nodes are 2 steps away, and the graph representation itself to tell you which nodes are one step away.  (Note that our example graph above doesn't actually have nodes at distance greater than 2 from each other.)

Some online sources have different formulas for closeness centrality, so please use our method:  find the average distance of all other vertices from the target vertex, then take the reciprocal of that average.

````python
def closeness_centrality(vertex, graph):
  two_away = friends_of_friends(vertex, graph)
  three_away = []
  one_away = []
  for node in two_away:
    for j in graph[node]:
      if j != vertex and j not in graph[vertex] and j not in three_away:
        three_away.append(j)
  for vert in graph[vertex]:
    one_away.append(vert)
  avg_dist = ((len(two_away) * 2) +(len(three_away) * 3) + len(one_away))/(len(two_away) + len(three_away) + len(one_away))
  return 1/avg_dist
````
